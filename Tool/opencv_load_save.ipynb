```python
import os
import cv2
import shutil

current_directory = os.getcwd()
print("Current Directory:", current_directory)

# List the contents of the current directory
current_list = os.listdir(current_directory)
print("Current List of Files and Directories:", "\n", current_list)
```

```python
# Input and output directories
# input_folder = 'earring_original/images/valid'
# output_folder = 'earring_opencv/images/valid'
# input_folder = 'earring_original/images/train'
# output_folder = 'earring_opencv/images/train'
input_folder = "earring_original_v2"
output_folder = "earring_opencv"


# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Check the number of files in a directory
print(len(os.listdir(input_folder)))
```

```python
# debugging: remove the hidden dot-underscore-files  eg: ._H58.jpg

error_load_img = []
success_load_img = 0
# Iterate through each file in the input folder
for file_name in os.listdir(input_folder):
    # Ensure the file is an image (you may want to add more checks)
    if file_name.lower().endswith((".png", ".jpg", ".jpeg")):
        # Read an image from file
        img = cv2.imread(os.path.join(input_folder, file_name))
        if img is not None:
            # Write the result_img to a file in the output folder with the same name
            output_file = os.path.join(output_folder, file_name)
            cv2.imwrite(output_file, img)
            success_load_img += 1
        else:
            print(f"Error loading image: {file_name}")
            error_load_img += file_name


print("image loaded successfully:", success_load_img)
print("image loaded erroneously:", len(error_load_img))
```

```python
# Define a list of image extensions
image_extensions = [".jpg", ".jpeg", ".png", ".bmp", ".JPG"]

# Create a list of image filenames in 'data_path'
imgs_list = [
    filename
    for filename in os.listdir(output_folder)
    if os.path.splitext(filename)[-1] in image_extensions
]

_ = [
    filename
    for filename in os.listdir(output_folder)
    if os.path.splitext(filename)[-1] not in image_extensions
]

print(imgs_list)
print("\n", len(imgs_list))

print(_)
print("\n", len(_))
```

```python
# Check Unique File Extension
base_names = []
file_extensions = []

for file_name in imgs_list:
    base_name, file_extension = file_name.split(".")
    base_names.append(base_name)
    file_extensions.append(file_extension)
    # print("Base Name:", base_name)
    # print("File Extension:", file_extension)
print(base_names)
print(file_extensions)

uni = list(set(file_extensions))
print(uni)
```

```python
# Replace the shuffled list of .jpg files with .txt files.
txt_list = []

for img in imgs_list:
    if img.split(".")[-1] in uni:
        txt_list.append(img.replace(img.split(".")[-1], "txt"))

print(txt_list)
print("\n", len(txt_list))
```

```python
# output_folder = 'earring_opencv/labels/valid'
# output_folder = "earring_opencv/labels/train"
output_folder = "earring_opencv"

# Create destination folders if they don't exist
for folder_path in [output_folder]:  # , test_folder]:
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
```

```python
# input_folder = 'earring_original/labels/valid'
# input_folder = "earring_original/labels/train"
input_folder = "earring_original_v2"

i = 0
# Copy txt files to destination folders
for f in txt_list:
    shutil.copy(os.path.join(input_folder, f), os.path.join(output_folder, f))
    i += 1

print("txt files copy successfully:", i)
```
