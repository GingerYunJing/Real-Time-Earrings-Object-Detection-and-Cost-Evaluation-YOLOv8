```python
import pyodbc
import pandas as pd
# %load_ext sql
print(pyodbc.version)
pyodbc.drivers()
```

```python
def db_conn(server = 'localhost', database = 'earring', username = '', password = ''):
    db = pyodbc.connect(f'Driver={{ODBC Driver 17 for SQL Server}};'
                        f'Server={server};'
                        f'database={database};'
                        f'UID={username};'
                        f'PWD={password};'
                        f'Trusted_Connection=yes;')
    cursor = db.cursor()
    return cursor, db
cursor, db = db_conn()

sql = 'select * from parts'
df=pd.read_sql(sql, db)
df.head()
```
